```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

### Preprocessing
- Read imputed and encoding data
- Split data by `split` feature

```python
df = pd.read_csv('../Week3/model_data_knn_after_encoding.csv.gz')
```

```python
meta_data = ["ClosePrice", "SaleMonth", "split", "CloseDate"]

feature_cols = [
    col for col in df.columns
    if col not in meta_data
]

# Restore sparse one-hot encoded zeros that were saved as blank cells in the CSV.
df[feature_cols] = (
    df[feature_cols]
    .fillna(0)
    .astype("float32")
)

print("Remaining NaN:", df[feature_cols].isna().sum().sum())

#Split Data
train = df[df["split"] == "train"]
test = df[df["split"] == "test"]



Y_train = train["ClosePrice"]
Y_test = test["ClosePrice"]

X_train = train.drop(columns = meta_data)
X_test = test.drop(columns=meta_data)

print(f"Train:\n{train.shape}")
print(X_train.shape)
print(Y_train.shape)
print("-----")
print(f"Test:\n{test.shape}")
print(X_test.shape)
print(Y_test.shape)
```

### Build Linear Model

```python
#building the linear regression model
lr_model = LinearRegression()

#fit training data
lr_model.fit(X_train, Y_train)

#Get prediction
Y_pred = lr_model.predict(X_test)

#Get model performance metric
r2 = r2_score(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)

print(f"R^2 Score: {r2}")
print(f"Mean Squared Error: {mse}")
```
